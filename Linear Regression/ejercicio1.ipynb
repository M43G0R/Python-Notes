```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("ingreso.csv", index_col=False)
```

```python
data.head(10)
```

La variable independiente va a ser las horas mientras que la variable dependiente es el ingreso y esta sera la eu debemos predecir.

```python
plt.figure(figsize=(10,8))
plt.scatter(data["horas"], data["ingreso"], color="black")
plt.xlabel('Horas')
plt.ylabel('Ingresos ($)')
plt.title('Gráfico de Dispersión')
plt.show()
```

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()#Aqui se crea un "creador de modelos" para hacer la regresion aun sin pasarle los datos

horas = data["horas"].values.reshape((-1,1))#Necesitamos pasar los datos que estan en la columna que estan en un mismo array a que pase a uno donde cada valor sea un array

modelo = regresion.fit(horas, data["ingreso"])#Primero ponemos los datos independientes y depues pasamos a los datos dependientes 

#Recordemos la formula para la recta y= mx + b
#Con los datos de las columnas y de la regresion ya podemos sacar los sig datos
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)


#Tambien podemos hacer prediccion automatizadas segun los datos que tenemos

entrada = [[39], [40], [43], [44]]
modelo.predict(entrada)#Aqui predecimos


plt.figure(figsize=(10,8))
plt.scatter(entrada, modelo.predict(entrada), color="red")
plt.plot(entrada, modelo.predict(entrada), color="blue")
plt.scatter(data["horas"], data["ingreso"], color="black")
plt.xlabel('Horas')
plt.ylabel('Ingresos ($)')
plt.title('Gráfico de Dispersión')
plt.show()

```
